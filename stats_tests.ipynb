```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import scipy.stats as stats
import numpy as np
from scipy.stats import sem
```

```python
file = "../ProjectOne_NationalParks/final_merge_data.csv"
data = pd.read_csv(file)
data.head()
```

```python
year = data.groupby(["Year", "Month"]).mean()
total_visitors = data.groupby(["Year", "Month"]).sum()
year = year.reset_index()
total_visitors = total_visitors.reset_index()
year["Total Visitors"] = total_visitors["Visits"]
del year["Lat"]
del year["Lng"]
del year["Visits"]
year
```

```python
sns.pairplot(year)
```

```python
a = year["Temp"]
b = year["Total Visitors"]
from scipy.stats import linregress
linregress(a, b)
```

```python
year_only = data.groupby(["Year"]).mean()
total_visitors_all = data.groupby(["Year"]).sum()
year_only = year_only.reset_index()
total_visitors_all = total_visitors_all.reset_index()
year_only["Total Visitors"] = total_visitors_all["Visits"]
del year_only["Lat"]
del year_only["Lng"]
del year_only["Visits"]
year_only
```

```python
a = year_only["Temp"]
b = year_only["Total Visitors"]
from scipy.stats import linregress
linregress(a, b)
```

```python
a = year_only["Year"]
b = year_only["Temp"]
from scipy.stats import linregress
linregress(a, b)
```

```python
a = year_only["Year"]
b = year_only["Total Visitors"]
from scipy.stats import linregress
linregress(a, b)
```

```python
group1 = year[year["Year"] == 2009]["Temp"]
group2 = year[year["Year"] == 2010]["Temp"]
group3 = year[year["Year"] == 2011]["Temp"]
group4 = year[year["Year"] == 2012]["Temp"]
group5 = year[year["Year"] == 2013]["Temp"]
group6 = year[year["Year"] == 2014]["Temp"]
group7 = year[year["Year"] == 2015]["Temp"]
group8 = year[year["Year"] == 2016]["Temp"]
group9 = year[year["Year"] == 2017]["Temp"]
group10 = year[year["Year"] == 2018]["Temp"]

stats.f_oneway(group1, group2, group3, group4, group5, group6, group7, group8, group9, group10)
```

```python
a = year_only["Year"]
b = year_only["Temp"]

#LinregressResult(slope=0.221551814100603, intercept=-393.5804253683318,
#rvalue=0.6939305568717172, pvalue=0.026011914723906977, stderr=0.08127782867198223)

plt.plot(a,b)
plt.ylim(min(b)-3,max(b)+3)

slope, int, r, p, std_err = stats.linregress(a, b)
fit = slope * a + int

plt.plot(a, fit, "k--", linewidth=1)
plt.style.use('seaborn')
plt.ylabel("Temperature (F)", fontsize=15)
plt.xlabel("Year", fontsize=15)
plt.title("Average Temperature (F) per Year", fontsize=18)
plt.xticks(a)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 8)
plt.figtext(.7, .2, "rvalue=0.69393 \npvalue=0.02601", horizontalalignment='left',
            fontsize=17, multialignment='left', bbox=dict(facecolor='white', edgecolor='black', pad=10.0))
plt.savefig('Average Temperature (F) per Year.png', dpi=300)

plt.show()
```
